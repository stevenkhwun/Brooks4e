```python
import pandas as pd
```

```python
# Specify values for each column
df1 = pd.DataFrame(
    {"a" : [4, 5, 6],
     "b" : [7, 8, 9],
     "c" : [10, 11, 12]},
    index = [1, 2, 3]
)
df1
```

```python
# Specify values for each row
df2 = pd.DataFrame(
    [[4, 7, 10],
     [5, 8, 11],
     [6, 9, 12]],
    index = [1, 2, 3],
    columns = ['a', 'b', 'c']
)
df2
```

```python
# Create DataFrame with a MultiIndex
df3 = pd.DataFrame(
    {"a" : [4, 5, 6],
     "b" : [7, 8, 9],
     "c" : [10, 11, 12]},
    index = pd.MultiIndex.from_tuples(
        [('d', 1), ('d', 2),
         ('e', 2)], names = ['n', 'v']
    )
)
df3
```
